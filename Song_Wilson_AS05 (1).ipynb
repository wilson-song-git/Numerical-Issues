```python
#Wilson Song
#Assignment 5

from math import*
from sympy import *
import numpy as np
from scipy.integrate import quad

#absolute error function
def absolute_error(exact,approx):
    return fabs(exact-approx)

#relative error function
def relative_error(exact,approx):
    return fabs(exact-approx)/fabs(exact)

#derivative of sin(x) = cos(x) so at x=0.9 it is cos(0.9)
x_variable=0.9

x=Symbol('x')
f=sin(x)
f_prime=f.diff(x)
f_prime=lambdify(x,f_prime)
exact_value= f_prime(x_variable)
print("\nExact Value = \n",str(exact_value))

h=0.1
three_point = ((np.sin(x_variable+h)-np.sin(x_variable-h)))/(2*h)
five_point = ((np.sin(x_variable-2*h)-8*np.sin(x_variable-h)+8*np.sin(x_variable+h))-np.sin(x_variable+2*h))/(12*h)

#calculate absolute and relative error for 3-point
absolute_error_3=absolute_error(exact_value,three_point)
relative_error_3=relative_error(exact_value,three_point)

#calculate absolute and relative error for 5-point
absolute_error_5=absolute_error(exact_value,five_point)
relative_error_5=relative_error(exact_value,five_point)

#print results and error of 3-point
print("\nResult using 3-point midpoint formula =\n",str(three_point))
print("\nAbsolute Error for 3-point =\n",absolute_error_3)
print("\nRelative Error for 3-point =\n",relative_error_3)

#print results and error of 3-point
print("\nResult using 5-point midpoint formula =\n",str(five_point))
print("\nAbsolute Error for 5-point =\n",absolute_error_5)
print("\nRelative Error for 5-point =\n",relative_error_5)



```

From this data we can see that using the 5-point midpoint formula provided value more accurate to the exact value than the 3-point midpoint formula as we expected.

```python
#Calculate sinx from 0.8 radians to 1.2 radians using the composite simpson's rule

#function sin(x)
def integrand(x):
    return np.sin(x)

#using scipy.integrate.quad to integrate and find the exact value of sin(x) from 0.8 radians to 1.2 radians
ans, err = quad(integrand,0.8,1.2)
exact=ans
print("\nExact Value =\n",str(exact))

#the composite simpson rule function
def composite_simpson_rule(f,initial,final,interval):
    
    #f = the function
    #intial is the lower bound
    #final is the upper bound
    #interval is the number of sub intervals - must be even
    
    h=(final-initial)/interval #step of integration
    k=0.0 # the sum
    x= initial+h
    for i in range(1,interval//2+1):
        k+=4*f(x)
        x+=2*h
        
    x=initial+2*h
    for i in range(1,interval//2):
        k+=2*f(x)
        x+=2*h
        
    return (h/3)*(f(initial)+f(final)+k)

#get the result for each subinterval 1,2,4,8
result_interval_1=composite_simpson_rule(integrand,0.8,1.2,1)
result_interval_2=composite_simpson_rule(integrand,0.8,1.2,2)
result_interval_4=composite_simpson_rule(integrand,0.8,1.2,4)
result_interval_8=composite_simpson_rule(integrand,0.8,1.2,8)

#Calculate relative and absolute error for 1 subinterval
relative_error_1=relative_error(exact,result_interval_1)
absolute_error_1=absolute_error(exact,result_interval_1)

#Calculate relative and absolute error for 2 subinterval
relative_error_2=relative_error(exact,result_interval_2)
absolute_error_2=absolute_error(exact,result_interval_2)

#Calculate relative and absolute error for 4 subinterval
relative_error_4=relative_error(exact,result_interval_4)
absolute_error_4=absolute_error(exact,result_interval_4)

#Calculate relative and absolute error for 8 subinterval
relative_error_8=relative_error(exact,result_interval_8)
absolute_error_8=absolute_error(exact,result_interval_8)

#printing out results and relative error for 1 subinterval
print("\nSimpson Rule with 1 subinterval =\n",str(result_interval_1))
print("\nAbsolute Error for 1 subinterval =\n",absolute_error_1)
print("\nRelative Error for 1 subinterval =\n",relative_error_1)

#printing out results and relative error for 2 subinterval
print("\nSimpson Rule with 2 subinterval =\n",str(result_interval_2))
print("\nAbsolute Error for 2 subinterval =\n",absolute_error_2)
print("\nRelative Error for 2 subinterval =\n",relative_error_2)

#printing out results and relative error for 4 subinterval
print("\nSimpson Rule with 4 subinterval =\n",str(result_interval_4))
print("\nAbsolute Error for 4 subinterval =\n",absolute_error_4)
print("\nRelative Error for 4 subinterval =\n",relative_error_4)

#printing out results and relative error for 8 subinterval
print("\nSimpson Rule with 8 subinterval =\n",str(result_interval_8))
print("\nAbsolute Error for 8 subinterval =\n",absolute_error_8)
print("\nRelative Error for 8 subinterval =\n",relative_error_8)
```

By analyzing the approximations of the integral from the composite simpson's rule and the error calculations, we can see that as we use more subintervals, the approximation is more accurate. This follows what we discussed in class in that we expected the error to decrease as we use more subintervals.
